```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://www.analytictech.com/mb021/mlk.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
speech = soup.find_all('p')
```

```python
paragraph_speech = [para.text for para in speech]
paragraph_speech
```

```python
string_speech = ' '.join(paragraph_speech)
string_speech
```

```python
clean_speech = string_speech.replace('\r\n', ' ')
clean_speech
```

```python
import re
```

```python
remove_punct_speech = re.sub(r'[^\w\s]', '' ,clean_speech)
remove_punct_speech
```

```python
speech_lower = remove_punct_speech.lower()
speech_lower
```

```python
speech_to_words = re.split(r'\s+', speech_lower)
speech_to_words
```

```python
import pandas as pd
```

```python
filtered_words = [word for word in speech_to_words if len(word) >= 4]
filtered_words
```

```python
df = pd.DataFrame(filtered_words).value_counts()
df
```

```python
df.to_csv(r'Enter your file path here\speech.csv', index_label = 'Words', header = ['Count'])
```
